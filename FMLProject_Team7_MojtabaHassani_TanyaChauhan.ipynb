```python
# 4: Loading the data set

# Set device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load dataset
ds = load_dataset("gOLIVES/OLIVES_Dataset", "biomarker_detection")
```

```python
# 5: using RESNET Model


# ==========================================================
#  RESNET18 + CLINICAL  •  OLIVES DATASET  (train / val / test)
# ==========================================================


# -------- 2.  LOAD OLIVES DATA --------------------------------------
ds = load_dataset("gOLIVES/OLIVES_Dataset", "biomarker_detection")
train_split = ds["train"]              # 3 556 samples
test_split  = ds["test"]               # 1 186 samples

# reserve part of train_split for validation
VAL_SIZE = 500                         # ← change if you like
TRAIN_SIZE = len(train_split) - VAL_SIZE

train_hf = train_split.select(range(TRAIN_SIZE))
val_hf   = train_split.select(range(TRAIN_SIZE, len(train_split)))

# -------- 3.  CUSTOM TORCH DATASET ----------------------------------
class OLIVESDataset(Dataset):
    def __init__(self, hf_split):
        self.data = hf_split
        self.tfm  = transforms.Compose([
            transforms.Resize((128, 128)),   # 64 64
            transforms.Grayscale(num_output_channels=3),
            transforms.ToTensor(),
        ])

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        ex = self.data[idx]

        labels = torch.tensor(
            [float(ex[f"B{i}"] or 0.0) for i in range(1, 7)], dtype=torch.float32
        )
        clinical = torch.tensor(
            [float(ex["BCVA"] or 0.0), float(ex["CST"] or 0.0)], dtype=torch.float32
        )

        img = ex["Image"] if isinstance(ex["Image"], Image.Image) \
             else Image.open(ex["Image"]).convert("L")
        oct_tensor = self.tfm(img)

        return {"oct": oct_tensor, "clinical": clinical, "label": labels}

# build PyTorch datasets / loaders
train_ds = OLIVESDataset(train_hf)
val_ds   = OLIVESDataset(val_hf)
test_ds  = OLIVESDataset(test_split)

train_loader = DataLoader(train_ds, batch_size=256, shuffle=True)
val_loader   = DataLoader(val_ds,   batch_size=32, shuffle=False)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False)

# -------- 4.  MODEL --------------------------------------------------
class MultiModalNet(nn.Module):
    def __init__(self):
        super().__init__()
        backbone = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        backbone.fc = nn.Identity()          # 512‑dim
        self.oct_model = backbone
        self.head = nn.Sequential(
            nn.Linear(512 + 2, 128),
            nn.ReLU(),
            nn.Linear(128, 6),
            nn.Sigmoid()
        )

    def forward(self, img, clin):
        feat = self.oct_model(img)
        x = torch.cat([feat, clin], dim=1)
        return self.head(x)

model = MultiModalNet().to(device)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# -------- 5.  TRAIN / VALIDATE --------------------------------------
EPOCHS = 1
train_loss = []   # holds average train loss per epoch
val_loss   = []   # holds average val   loss per epoch
for epoch in range(1, EPOCHS + 1):
    # --- train ---
    model.train()
    epoch_train_loss = 0.0 # Initialize an accumulator for the epoch's loss
    for batch in tqdm(train_loader, desc=f"Train {epoch}/{EPOCHS}"):
        img = batch["oct"].to(device)
        cli = batch["clinical"].to(device)
        lab = batch["label"].to(device)

        optimizer.zero_grad()
        loss = criterion(model(img, cli), lab)
        loss.backward()
        optimizer.step()
        epoch_train_loss += loss.item() # Accumulate batch loss

    epoch_train_loss /= len(train_loader) # Calculate average epoch loss

    # --- validate ---
    model.eval()
    epoch_val_loss = 0.0 # Similar accumulator for validation loss
    with torch.no_grad():
        for batch in val_loader:
            img = batch["oct"].to(device)
            cli = batch["clinical"].to(device)
            lab = batch["label"].to(device)
            epoch_val_loss += criterion(model(img, cli), lab).item()
    epoch_val_loss /= len(val_loader)


    train_loss.append(epoch_train_loss) # Append average epoch losses to the list
    val_loss.append(epoch_val_loss)

    print(f"Epoch {epoch:02d} | Train loss {epoch_train_loss:.4f} | Val loss {epoch_val_loss:.4f}")

# -------- 6.  TEST ---------------------------------------------------
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for batch in tqdm(test_loader, desc="Test"):
        img = batch["oct"].to(device)
        cli = batch["clinical"].to(device)
        preds = model(img, cli).cpu()
        all_preds.append(preds)
        all_labels.append(batch["label"])

all_preds  = torch.cat(all_preds).numpy()
all_labels = torch.cat(all_labels).numpy()
print("✅ Evaluation on full test set complete.")

```

```python
# 13: Cross validation to compare ResNet with VGG
# but due to time limits, we could not finish this block
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.model_selection import KFold
from datasets import load_dataset
from PIL import Image
import numpy as np
from tqdm import tqdm

# Load dataset
ds = load_dataset("gOLIVES/OLIVES_Dataset", "biomarker_detection")
full_data = ds["train"]
test_data = ds["test"]

# Custom dataset class
class OLIVESDataset(Dataset):
    def __init__(self, hf_split):
        self.data = hf_split
        self.tfm = transforms.Compose([
            transforms.Resize((32, 32)),
            transforms.Grayscale(num_output_channels=3),
            transforms.ToTensor(),
        ])

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        ex = self.data[idx]
        labels = torch.tensor([
            float(ex[f"B{i}"] or 0.0) for i in range(1, 7)
        ], dtype=torch.float32)
        clinical = torch.tensor([
            float(ex["BCVA"] or 0.0), float(ex["CST"] or 0.0)
        ], dtype=torch.float32)

        img = ex["Image"] if isinstance(ex["Image"], Image.Image) \
              else Image.open(ex["Image"]).convert("L")
        oct_tensor = self.tfm(img)

        return {"oct": oct_tensor, "clinical": clinical, "label": labels}

# Model class
class MultiModalNet(nn.Module):
    def __init__(self, backbone_name="resnet18"):
        super().__init__()
        if backbone_name == "resnet18":
            backbone = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        elif backbone_name == "vgg16":
            backbone = models.vgg16(weights=models.VGG16_Weights.DEFAULT)
            backbone.classifier = nn.Identity()
            backbone.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        else:
            raise ValueError("Unsupported backbone")

        if "resnet" in backbone_name:
            backbone.fc = nn.Identity()
            self.feature_dim = 512
        else:
            self.feature_dim = 512 * 7 * 7

        self.oct_model = backbone
        self.head = nn.Sequential(
            nn.Linear(self.feature_dim + 2, 128),
            nn.ReLU(),
            nn.Linear(128, 6),
            nn.Sigmoid()
        )

    def forward(self, img, clin):
        feat = self.oct_model(img)
        if len(feat.shape) > 2:
            feat = torch.flatten(feat, 1)
        x = torch.cat([feat, clin], dim=1)
        return self.head(x)

# Settings
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
EPOCHS = 5
kf = KFold(n_splits=5, shuffle=True, random_state=42)
backbones = ["resnet18", "vgg16"]

# Final predictions and losses
final_scores = {b: [] for b in backbones}

# Cross-validation
for backbone in backbones:
    print(f"\n======= Evaluating {backbone.upper()} =======")
    all_fold_preds = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(full_data)):
        print(f"\nFold {fold+1}/5")

        train_hf = full_data.select(train_idx.tolist())
        val_hf = full_data.select(val_idx.tolist())

        train_ds = OLIVESDataset(train_hf)
        val_ds = OLIVESDataset(val_hf)
        test_ds = OLIVESDataset(test_data)

        train_loader = DataLoader(train_ds, batch_size=256, shuffle=True)
        val_loader = DataLoader(val_ds, batch_size=32, shuffle=False)
        test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)

        model = MultiModalNet(backbone_name=backbone).to(device)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=1e-4)

        for epoch in range(1, EPOCHS + 1):
            model.train()
            for batch in tqdm(train_loader, desc=f"Train Epoch {epoch}"):
                img, cli, lab = batch["oct"].to(device), batch["clinical"].to(device), batch["label"].to(device)
                optimizer.zero_grad()
                loss = criterion(model(img, cli), lab)
                loss.backward()
                optimizer.step()

        model.eval()
        fold_preds = []
        with torch.no_grad():
            for batch in test_loader:
                img, cli = batch["oct"].to(device), batch["clinical"].to(device)
                preds = model(img, cli).cpu()
                fold_preds.append(preds)

        all_fold_preds.append(torch.cat(fold_preds, dim=0))

    # Average predictions over folds
    avg_preds = torch.stack(all_fold_preds).mean(dim=0)
    final_scores[backbone] = avg_preds

# Compare models
print("\n========== FINAL COMPARISON ==========\n")
for backbone in backbones:
    avg_conf = final_scores[backbone].mean().item()
    print(f"{backbone.upper():>8} | Avg Confidence: {avg_conf:.4f}")





```
